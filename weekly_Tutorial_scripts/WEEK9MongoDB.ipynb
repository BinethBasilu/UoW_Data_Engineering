```python
import pymongo
from pymongo import MongoClient
```

```python
Image_connection = pymongo.MongoClient("mongodb+srv://binethmunaweera:mkbinethbasilu@weeknine.5xvii.mongodb.net/?retryWrites=true&w=majority&appName=weeknine")
db = Image_connection["CBIR"]
```

```python
try:
  print(db.list_collection_names())
except Exception as e:
  print(f"Connection failed: {e}")
```

```python
image_metadata_collection = db["image_metadata"]
```

```python
import json
with open('/content/image_metadata.json') as file:
  image1_data = json.load(file)

if isinstance(image1_data, list):
  image_metadata_collection.insert_many(image1_data)
else:
  image_metadata_collection.insert_one(image1_data)
```

```python
processed_collection = db["processed_images"]
```

```python
import os
from bson import Binary

image_folder_path = '/content/images'
for filename in os.listdir(image_folder_path):
  if filename.endswith('.png') or filename.endswith('.jpg'):
    with open(os.path.join(image_folder_path, filename), 'rb') as image_file:
      binary_image = Binary(image_file.read())
      image_doc = {
          "filename": filename,
          "image": binary_image
      }
      processed_collection.insert_one(image_doc)
```

```python
query = {"filename": "Image12.jpg"}
image_document = processed_collection.find_one(query)
if image_document:
  print("Image found:", image_document["filename"])
else:
  print("No image found with the specific criteria.")
```

```python
text_connection = pymongo.MongoClient("mongodb+srv://binethmunaweera:mkbinethbasilu@weeknine.5xvii.mongodb.net/?retryWrites=true&w=majority&appName=weeknine")
db = text_connection["NLP"]
```

```python
try:
  print(db.list_collection_names())
except Exception as e:
  print(f"Connection failed: {e}")
```

```python
doc_metadata_collection = db["doc_metadata"]
```

```python
import json
with open ('/content/text_metadata.json') as file:
  doc1_data = json.load(file)
  if isinstance(doc1_data, list):
    doc_metadata_collection.insert_many(doc1_data)
  else:
    doc_metadata_collection.insert_one(doc1_data)
```

```python
processed_doc_collection = db["Processed_doc"]
```

```python
import os
from datetime import datetime

document_folder_path = '/content/Documents'
for filename in os.listdir(document_folder_path):
  file_path = os.path.join(document_folder_path, filename)
  if os.path.isfile(file_path):
    with open(file_path, 'r') as file:
      file_content = file.read()
      document_data = {
          "filename": filename,
          "content": file_content,
          "metadata":{
              "processed_date": datetime.utcnow().isoformat(),
              "tags": ["tags","document"]
          }
      }
      processed_doc_collection.insert_one(document_data)
```

```python
query = {"filename": "text20"}
text_document = processed_doc_collection.find_one(query)
if text_document:
  print("Doc found:", text_document["filename"])
else:
  print("No text doc found with the specific criteria.")
```
